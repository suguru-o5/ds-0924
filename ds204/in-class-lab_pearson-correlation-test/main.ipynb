## Part 1: Data Familiarization

```python
data = pd.read_csv("Advertising.csv")

# check first 10 observations
data.head(10)
```

## Part 2: Pearson Correlation Analysis

#### 1. Select Variables:

Choose one pair of variables to analyze first:
 - TV Advertising and Sales

```python
# Calculate the Pearson correlation coefficient between TV advertising and Sales
tv_sales_corr = data['TV'].corr(data['Sales'])
print(f'Pearson correlation between TV and Sales: {tv_sales_corr}')
```

## Part 3: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot
sns.scatterplot(x=data['TV'], y=data['Sales'])

# Add a trendline
sns.regplot(x=data['TV'], y=data['Sales'], scatter=False, color='red')

# Add title and labels
plt.title('Relationship between TV Advertising and Sales')
plt.xlabel('TV Advertising Budget (USD)')
plt.ylabel('Sales (USD)')
plt.show()
```

```python
# Calculate the Pearson correlation coefficient between Radio advertising and Sales
radio_sales_corr = data['Radio'].corr(data['Sales'])
print(f'Pearson correlation between Radio and Sales: {radio_sales_corr}')

# Create a scatter plot for Radio and Sales
sns.scatterplot(x=data['Radio'], y=data['Sales'])
sns.regplot(x=data['Radio'], y=data['Sales'], scatter=False, color='red')
plt.title('Relationship between Radio Advertising and Sales')
plt.xlabel('Radio Advertising Budget (USD)')
plt.ylabel('Sales (USD)')
plt.show()
```

```python
# Calculate the Pearson correlation coefficient between Newspaper advertising and Sales
newspaper_sales_corr = data['Newspaper'].corr(data['Sales'])
print(f'Pearson correlation between Newspaper and Sales: {newspaper_sales_corr}')

# Create a scatter plot for Newspaper and Sales
sns.scatterplot(x=data['Newspaper'], y=data['Sales'])
sns.regplot(x=data['Newspaper'], y=data['Sales'], scatter=False, color='red')
plt.title('Relationship between Newspaper Advertising and Sales')
plt.xlabel('Newspaper Advertising Budget (USD)')
plt.ylabel('Sales (USD)')
plt.show()
```
