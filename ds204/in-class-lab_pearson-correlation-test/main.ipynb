# In-Class Lab: Pearson Correlation Test Using Real World Advertising Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Part 1: Data Familiarization

```python
data = pd.read_csv("Advertising.csv")

# check first 10 observations
data.head(10)
```

3. Identify research questions:
- Is there a significant relationship between TV advertising and sales?

## Part 2: Pearson Correlation Analysis

#### 1. Select Variables:

Choose one pair of variables to analyze first:
 - TV Advertising and Sales

```python
# Calculate the Pearson correlation coefficient between TV advertising and Sales
tv_sales_corr, tv_sales_p = stats.pearsonr(data['TV'], data['Sales'])
print(f"Pearson correlation between TV and Sales: {tv_sales_corr}")
print(f"P-Value: {tv_sales_p}")
```

- If r is close to 1, it indicates a strong positive linear relationship.
- If r is close to -1, it indicates a strong negative relationship.
- If r is close to 0, there is little or no linear relationship.

Here we have 0.7822244248616066 which indicates that a strong positive linear relationship.  
Also, p-value which is less than 0.05 indicates that this result is statistically significant.

## Part 3: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot
sns.scatterplot(x=data['TV'], y=data['Sales'])

# Add a trendline
sns.regplot(x=data['TV'], y=data['Sales'], scatter=False, color='red')

# Add title and labels
plt.title('Relationship between TV Advertising and Sales')
plt.xlabel('TV Advertising Budget (USD)')
plt.ylabel('Sales (USD)')
plt.show()
```

- The scatterplot shows a strong positive linear relationship between TV advertising and sales, with data points closely following the trend line.
- The correlation coefficient (r) from the earlier calculation is 0.78, which supports this strong relationship.

```python
# Calculate the Pearson correlation coefficient between Radio advertising and Sales
radio_sales_corr, radio_sales_p = stats.pearsonr(data['Radio'], data['Sales'])
print(f"Pearson correlation between Radio and Sales: {radio_sales_corr}")
print(f"P-Value: {radio_sales_p}")

# Create a scatter plot for Radio and Sales
sns.scatterplot(x=data['Radio'], y=data['Sales'])
sns.regplot(x=data['Radio'], y=data['Sales'], scatter=False, color='red')
plt.title('Relationship between Radio Advertising and Sales')
plt.xlabel('Radio Advertising Budget (USD)')
plt.ylabel('Sales (USD)')
plt.show()
```

- The scatterplot indicates a moderate to strong positive linear relationship. While the trend line captures the general trend, the spread of data points is slightly wider than for TV.  
- This suggests a slightly weaker correlation compared to TV.

```python
# Calculate the Pearson correlation coefficient between Newspaper advertising and Sales
newspaper_sales_corr, newspaper_sales_p = stats.pearsonr(data['Newspaper'], data['Sales'])
print(f"Pearson correlation between Newspaper and Sales: {newspaper_sales_corr}")
print(f"P-Value: {newspaper_sales_p}")

# Create a scatter plot for Newspaper and Sales
sns.scatterplot(x=data['Newspaper'], y=data['Sales'])
sns.regplot(x=data['Newspaper'], y=data['Sales'], scatter=False, color='red')
plt.title('Relationship between Newspaper Advertising and Sales')
plt.xlabel('Newspaper Advertising Budget (USD)')
plt.ylabel('Sales (USD)')
plt.show()
```

- The scatterplot shows a weak positive linear relationship, with the data points scattered more widely around the trend line.
- This weaker relationship implies that newspaper advertising has the least impact on sales compared to TV and radio.
