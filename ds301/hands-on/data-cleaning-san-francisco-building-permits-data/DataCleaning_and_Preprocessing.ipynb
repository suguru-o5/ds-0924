### Mean and Median Imputation
#### Both mean and median imputation are techniques used to handle missing data by replacing missing values with a central tendency measure of the available data.

*1. Mean Imputation*
*Definition:* Missing values are replaced with the mean (average) of the non-missing values in the column.
*When to Use:*
When data is normally distributed (symmetrical).
When you want to preserve the mean of the dataset.
Works well for continuous numerical features.
*Limitations:*
Affected by outliers, which can skew the imputation.
Reduces variance in the dataset, leading to loss of variability.
*2. Median Imputation*
*Definition:* Missing values are replaced with the median (middle value) of the non-missing values.
*When to Use:*
When data is skewed or contains outliers.
Works well for continuous and ordinal data.
Helps in preserving the distribution of the data better than mean imputation.
*Limitations:*
Does not consider relationships with other variables.
Can still distort variability but less than mean imputation.

```python
#Performing mean imputation on Estimated Cost
data['Estimated Cost'] = data['Estimated Cost'].fillna(data['Estimated Cost'].mean())
```

```python
#Performing median imputation on Revised Cost
data['Revised Cost'] = data['Revised Cost'].fillna(data['Revised Cost'].median())
```

### Using ffill() and bfill()
Forward fill (ffill()) fills missing values using the previous available value in the same group.
In this case, This assumes that if a building's permit was previously recorded, the same structure details apply.
Backward fill (bfill()) fills missing values using the next available value in the same group.
This helps when the first row(s) of a group have missing values (since ffill() can't fill upwards).

### When Not to Use Both?
If missing values are random (not grouped by address), ffill() and bfill() may introduce errors.
If missing values are too common, consider median imputation instead.

```python
#Handling missing values in Location using the Geopy
from geopy.geocoders import Nominatim
import time

geolocator = Nominatim(user_agent="sf_permit_locator")

def geocode_location(row):
    if pd.isna(row['Latitude']) or pd.isna(row['Longitude']):
        address = f"{row['Street Number']} {row['Street Name']}, San Francisco, CA {row['Zipcode']}"
        try:
            location = geolocator.geocode(address, timeout=10)
            if location:
                return pd.Series([location.latitude, location.longitude])
        except Exception as e:
            print(f"Geocoding failed for {address}: {e}")
            return pd.Series([None, None])
    return pd.Series([row['Latitude'], row['Longitude']])
```

```python
data[['Latitude', 'Longitude']] = data.apply(geocode_location, axis=1)
```
